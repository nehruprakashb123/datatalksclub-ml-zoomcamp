```python
X = df_test.values
XTX = X.T.dot(X)
XTX_inv = np.linalg.inv(XTX)
XTX_inv
```

```python
y = np.array([1100, 1300, 800, 900, 1000, 1100, 1200])
```

```python
w = (XTX_inv @ X.T) @ y
```
